```python
data_url = 'https://drive.google.com/uc?export=download&id=1mCjtYHiX--mMUjicuaP2gH3k-SnFxt8Y'
import pandas as pd
data = pd.read_csv(data_url, index_col=0, sep=', ', engine='python')
data.head()
```

```python
import plotly.figure_factory as ff


colors = ['#333F44', '#37AA9C']

fig = ff.create_distplot((data.iloc[:, 0], data.iloc[:, 1]), data.columns, show_hist=False, colors=colors)

fig.update_layout(title_text='Blood Pressure')
fig.show()
```

Yes we can clearly see that there is an decrease in average blood pressure.
